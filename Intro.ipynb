```python
import re
import os
import xml.etree.ElementTree as ET
import mysql.connector 
import pandas as pd
```

```python
def lectura_xml(ruta):
    

    tree = ET.parse(ruta)
    root = tree.getroot()

    xml_lista = []
    for child in root:
        xml_dict = {}
        for subchild in child:
            xml_dict[subchild.tag] = subchild.text
            xml_lista.append(xml_dict) # Es una lista que contiene diccionarios
        return xml_lista
```

```python
lectura_xml("Datos/data_xml_clean.xml")
```
