```python

m = folium.Map(location=[15.5,	49.75], tiles="OpenStreetMap", zoom_start=2.5,  control_scale=False)
```

```python
Man= df_mapa_freq[df_mapa_freq['genero']=='Man']
```

```python
Woman= df_mapa_freq[df_mapa_freq['genero']=='Woman']
```

```python


for i in range(0,len(Woman)):
   folium.Circle(
      location=[Woman.iloc[i]['lat'], Woman.iloc[i]['lon']],
      popup= [Woman.iloc[i]['pais'],  Woman.iloc[i]['count'], 'mujeres-data'],
      radius=float(Woman.iloc[i]['count'])*400,
      color='pink',
      fill=False,
      fill_color='pink'
   ).add_to(m)     

   for i in range(0,len(Man)):
   folium.Circle(
      location=[Man.iloc[i]['lat'], Man.iloc[i]['lon']],
      popup= [Man.iloc[i]['pais'],  Man.iloc[i]['count'], 'hombres-data'],
      radius=float(Man.iloc[i]['count'])*400,
      color='blue',
      fill=False,
      fill_color='blue'
   ).add_to(m)      
m 

```
