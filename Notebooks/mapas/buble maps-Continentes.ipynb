```python
import folium
import pandas as pd
import sidetable as stb
```

```python
data = pd.DataFrame({
   'lon':[-58.567698,21.401132, 15.424569, 86.440194, 133.989279, -106.503336],
   'lat':[-7.102691, 10.750952,51.236945, 39.845398, -13.457803, 49.030387],
   'continente':['América del Sur', 'África', 'Europa', 'Asia', 'Oceanía', 'América del Norte'],

}, dtype=str)

data
```

```python
df_mapa= pd.read_csv ('social_y_coordenadas.csv', index_col=0)
```

```python
df_social= pd.read_csv ('df-social-limpio.csv', index_col=0)
```

```python
df_mapa_freq= df_mapa.stb.freq(['continente', 'genero'])
```

```python
df_mapa_freq
```

```python
df_mapa_freq.head(1)
```

```python
df_social_coordenadas= pd.merge (on='continente', how='inner', left=df_mapa_freq , right=data)
```

```python
df_social_coordenadas.head(2)
```

```python
Man= df_social_coordenadas[df_social_coordenadas['genero']=='Man']
Man
```

```python
Woman= df_social_coordenadas[df_social_coordenadas['genero']=='Woman']
Woman
```

```python

m = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)
```

```python

# hombres
for i in range(0,len(Man)):
   folium.Circle(
      location=[Man.iloc[i]['lat'], Man.iloc[i]['lon']],
      popup= [Man.iloc[i]['continente'],  Man.iloc[i]['count'], 'hombres-data'],
      radius=float(Man.iloc[i]['count'])*300,
      color='blue',
      fill=False,
      fill_color='blue'
   ).add_to(m)      
m
```

```python
#mujeres
for i in range(0,len(data)):
   folium.Circle(
      location=[Woman.iloc[i]['lat'], Woman.iloc[i]['lon']],
      popup= [Woman.iloc[i]['continente'],  Woman.iloc[i]['count'], 'mujeres-data'],
      radius=float(Woman.iloc[i]['count'])*300,
      color='pink',
      fill=True,
      fill_color='blue'
   ).add_to(m)

m
```
